```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('Diseases - Sheet1 (1).csv')
df.head()
```

```python
symp = []
# append different male and female symptom
for i in range(df.shape[0]):
    a = df['General_Symptom'].values[i].split(',')
    for i in a:
        tmp = i.strip()
        symp.append(tmp)
    
```

```python
symptom = []
for i in symp:
    if i not in symptom:
        temp = {}
        temp['name'] = i
        symptom.append(temp)
```
